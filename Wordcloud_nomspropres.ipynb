```python
import pandas as pd
data = pd.read_csv("/users/margotpascual/Documents/CorpusLEM/Verne_Vingt_Mille_Lieues_Sous_Les_Mers.tsv", sep="\t")
data.head()
```

```python
#on garde uniquement les entités nommées Noms Propres

data = data.loc[data['POS'] == 'NOMpro']
```

```python
data
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
mots = " ".join([mot for mot in data["treated"]])
```

```python
mots
```

```python
nuage_mots = WordCloud(background_color="white").generate(mots)
```

```python
#on crée le nuage de mots des noms propres

plt.figure(figsize=(12,7))
plt.imshow(nuage_mots)
```

```python
import nltk
nltk.download("punkt")
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
mots
```
