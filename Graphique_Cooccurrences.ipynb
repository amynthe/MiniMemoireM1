```python
import matplotlib.pyplot as plt
import csv
  
x = []
y = []
  
with open('Co-occurencesessai.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    
    header = next(plots)
    print(header) 
    
    for row in plots:
        x.append(row[0])
        y.append(int(row[1]))  

  
plt.bar(x, y, color = 'g', width = 0.72, label = "Co-Occurences")
plt.xlabel(header[0]) # on met le header comme label
plt.ylabel(header[1])
plt.xticks(rotation=90) # on tourne le texte des x à 90degré pour plus de lisibilité)
plt.title('Co-Occurences des différents personnages')
plt.legend()
plt.show()
```
